#### Use the appropriate plot to display the summary statistics of the `Age` in function of the `Pclass`.

*Hint: use boxplot to display summary statistics of `Age` in relation to each of the discrete values of `Pclass`*

```python
#your code here
clas1=pd.DataFrame(titanic[titanic['Pclass']==1])
clas2=pd.DataFrame(titanic[titanic['Pclass']==2])
clas3=pd.DataFrame(titanic[titanic['Pclass']==3])
fig,axes = plt.subplots(figsize = (16,4), nrows = 1, ncols = 3)

plt.sca(axes[0])
plt.title('Stats in Class 1')
plt.boxplot(clas1['Age'],vert=False);
plt.sca(axes[1])
plt.title('Stats in Class 2')
plt.boxplot(clas2['Age'],vert=False);
plt.sca(axes[2])
plt.title('Stats in Class 3')
plt.boxplot(clas3['Age'],vert=False);
```

#### Use the appropriate plot to display the distributions of  `Age` in relation to  `Gender`.

*Hint: use Facetgrid to display the distribution of `Age` for each categorical value of `Gender`*

```python
#your code here
mas=pd.DataFrame(titanic[titanic['Gender']=='male'])
fem=pd.DataFrame(titanic[titanic['Gender']=='female'])
fig,axes = plt.subplots(figsize = (16,4), nrows = 1, ncols = 2)

plt.sca(axes[0])
plt.title('Distribution of Age by males')
sns.distplot(mas.Age);
plt.sca(axes[1])
plt.title('Distribution of Age by females')
sns.distplot(fem.Age);
```
