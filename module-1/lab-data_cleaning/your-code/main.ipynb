#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
outliers=pd.DataFrame(columns=df4.columns)
```

```python
outliers
```

```python
stats=df4.describe().T
stats['IQR']=stats['75%']-stats['25%']
stats
```

```python
for m in stats.index:
    iqr=stats.at[m, 'IQR']
    cutoff=iqr*1.5  # umbral
    lower=stats.at[m, '25%']-cutoff
    upper=stats.at[m, '75%']+cutoff
    results=df4[(df4[m]<lower)|(df4[m]>upper)].copy()
    results['outlier']=m
    outliers=outliers.append(results, sort=True)
```

```python
outliers['outlier'].value_counts()
```

```python
outliers.to_csv('outliers.csv')
```

```python
clean_index=[i for i in df4.index if i not in outliers.index]
df4superclean=pd.DataFrame(columns=df4.columns)
df4superclean=df4.iloc[clean_index,:]
print (df4.shape, df4superclean.shape)
```
